# Tạo data train và vocabulary

```python
import os
file_list = []
for (dirpath, dirname, filename) in os.walk('Train_Full/Phap luat'):
    for f in filename:
        file_list.append(dirpath+'/'+f)
len(file_list)
```

```python
import string
import re, string
from tqdm import tqdm
from pyvi import ViTokenizer
from keras.preprocessing.text import Tokenizer

data = ""
tokenizer = Tokenizer()
for f in tqdm(file_list):
    f1 = open(f, encoding="utf-16")
    doc = f1.read()
    doc = ViTokenizer.tokenize(doc)
    #Lower
    doc= doc.lower()
    
    #prepare regex for char filtering
    exclude = set(string.punctuation.replace("_", ""))
    data+= ''.join(ch for ch in doc if ch not in exclude) 
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
```

```python
tokenizer.word_index
```

```python
encoded = tokenizer.texts_to_sequences([data])[0]

sequences = list()
for i in range(2, len(encoded)):
    sequence = encoded[i-2:i+1]
    sequences.append(sequence)
```

```python
import pickle
with open('sequences_digit_VH.pkl', 'wb') as f1:
            pickle.dump(sequences, f1)
        
with open('tokenizer_VH.pkl', 'wb') as f2:
            pickle.dump(tokenizer, f2)
```
